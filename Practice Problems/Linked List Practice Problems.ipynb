## Question 1 

Write a function that moves the last node to the front in a given Singly Linked List. <br>
Examples: <br>
Input: 1 → 2 → 3 → 4 → 5 <br>
Output: 5 → 1 → 2 → 3 → 4 <br>
Input: 3 → 8 → 1 → 5 → 7 → 12 <br>
Output: 12 → 3 → 8 → 1 → 5 → 7 <br>

```python
class Node:
    
    def __init__(self, elem, next):
        self.element = elem
        self.next = next


class SingleLinkedList:
    
    def __init__(self, a):
        if type(a) is list:
            self.head = Node(a[0], None)
            tail = self.head
            for i in range(1, len(a)):
                new_node = Node(a[i], None)
                tail.next = new_node
                tail = new_node
        else:
            self.head = a
    
    def printList(self):
        temp = self.head
        while temp.next != None:
            print(temp.element, end=" → ")
            temp = temp.next
        print(temp.element)
        return
    
    def question1(self):
        temp = self.head
        while temp.next.next != None:
            temp = temp.next
        last_node = temp.next
        temp.next = None
        last_node.next = self.head
        self.head = last_node
        return self.head
```

You are given a linked list that contains N integers. You have performed the following <br>
reverse operation on the list: <br>
Select all the subparts of the list that contain only even integers. For example, if the list <br>
is {1, 2, 8, 9, 12, 16}, then the selected subparts will be {2, 8}, {12, 16}. <br>
Reverse the selected subpart such as {8, 2} and {16, 12}. <br>
Now, you are required to retrieve the original list. <br>
Input format <br>
First line: N <br>
Next line: N space-separated integers that denote elements of the reverse list <br>
Output format <br>
Print the N elements of the original list. <br>
Sample Input <br>
9 <br>
2 18 24 3 5 7 9 6 12 <br>
Sample Output <br>
24 18 2 3 5 7 9 12 6 <br>
Explanation <br>
In the sample, the original list is {24, 18, 2, 3, 5, 7, 9, 12, 6,} which when reversed <br>
according to the operations will result in the list given in the sample input. <br>

```python
class Node:
    
    def __init__(self, element=None, next=None):
        self.element = element
        self.next = next


class SingleLinkedList:
    
    def __init__(self, a):
        if type(a) is list:
            self.head = Node(a[0], None)
            tail = self.head
            for i in range(1, len(a)):
                new_node = Node(a[i], None)
                tail.next = new_node
                tail = new_node
        else:
            self.head = a
    
    def printList(self):
        temp = self.head
        while temp.next != None:
            print(temp.element, end=" → ")
            temp = temp.next
        print(temp.element)
        return
    
    def reverse(self, before_first=None, after_last=None):
        prev = after_last
        current = before_first.next if before_first else self.head
        while current and current != after_last:
            prev, current.next, current = current, prev, current.next
        if before_first:
            before_first.next = prev
        else:
            self.head = prev
            
    def reverse_sub_parts(self):
        prev = None
        current = self.head
        while current:
            if current.element % 2 == 0:
                if not prev or prev.element % 2 == 1:
                    before_start = prev
            elif prev and prev.element % 2 == 0:
                self.reverse(before_start, current)
            prev = current
            current = current.next

        if prev and prev.element % 2 == 0:
            self.reverse(before_start)
```

Given a linked list, check if the linked list has a loop or not. The below diagram shows a linked <br>

### Question 09


A number is Palindrome if it reads same from front as well as back.For example, <br>
2332 is palindrome number as its read same from both sides. <br>
Linked List can also be palindrome if they have the same order when it traverse from forward as <br>
well as backward. <br>
Write a function that will take a linked list as input and return True if the list is a palindrome and <br>
return False otherwise. <br>
Sample Input 1: <br>
1 → 2 → 3 → 2 → 1 → None <br>
Sample Output 1: <br>
True <br>
Sample Input 2: <br>
1 → 2 → 3 → 1 → 1 → None <br>
Sample Output 2: <br>
False <br>

```python
class Node:
    
    def __init__(self, element=None, next=None):
        self.element = element
        self.next = next


class SingleLinkedList:
    
    def __init__(self, a):
        if type(a) is list:
            self.head = Node(a[0], None)
            tail = self.head
            for i in range(1, len(a)):
                new_node = Node(a[i], None)
                tail.next = new_node
                tail = new_node
        else:
            self.head = a
    
    def printList(self):
        temp = self.head
        while temp.next != None:
            print(temp.element, end=" → ")
            temp = temp.next
        print(temp.element)
        return
    
    def countNode(self):
        c = 0
        temp = self.head
        
        while temp:
            c += 1
            temp = temp.next
            
        return c
            
    
    def nodeAt(self, idx):
        length = self.countNode()
        if idx > length:
            return None
        else:
            start = self.head
            for i in range(idx):
                start = start.next
            return start
    
    def isPalindrome(self):
        head = self.head
    # rev records the first half, need to set the same structure as fast, slow, hence later we have rev.next
        rev = None
    # initially slow and fast are the same, starting from head
        slow = fast = head
        while fast and fast.next:
        # fast traverses faster and moves to the end of the list if the length is odd
            fast = fast.next.next
        
        # take it as a tuple being assigned (rev, rev.next, slow) = (slow, rev, slow.next), hence the re-assignment of slow would not affect rev (rev = slow)
            rev, rev.next, slow = slow, rev, slow.next
        if fast:
       # fast is at the end, move slow one step further for comparison(cross middle one)
            slow = slow.next
    # compare the reversed first half with the second half
        while rev and rev.element == slow.element:
            slow = slow.next
            rev = rev.next
    
    # if equivalent then rev become None, return True; otherwise return False 
        return not rev
```

### Question - 10

```python
class Node:
    
    def __init__(self, element=None, next=None):
        self.element = element
        self.next = next


class SingleLinkedList:
    
    def __init__(self, a):
        if type(a) is list:
            self.head = Node(a[0], None)
            tail = self.head
            for i in range(1, len(a)):
                new_node = Node(a[i], None)
                tail.next = new_node
                tail = new_node
        else:
            self.head = a
    
    def printList(self):
        temp = self.head
        while temp.next != None:
            print(temp.element, end=" → ")
            temp = temp.next
        print(temp.element)
        return
    
    def countNode(self):
        c = 0
        temp = self.head
        
        while temp:
            c += 1
            temp = temp.next
            
        return c
            
    
    def nodeAt(self, idx):
        length = self.countNode()
        if idx > length:
            return None
        else:
            start = self.head
            for i in range(idx):
                start = start.next
            return start
        
        
    def remove(self, idx):
        count = self.countNode()
        if idx < 0 or idx > count:
            raise Exception("Invalid index")
        
        if idx == 0:
            temp = self.head
            temp1 = self.head.next
            self.head = temp1
            return temp.element
        
        elif idx == count:
            temp = self.nodeAt(count)
            temp1 = self.nodeAt(idx-1)
            temp1.next = None
            return temp.element
        else:
            temp = self.nodeAt(idx-1)
            temp1 = self.nodeAt(idx+1)
            temp2 = self.nodeAt(idx)
            temp.next = temp1
            return temp2.element
            
     
    def question10(self):
        head = self.head
        if (head == None):
            return
  
    # Initialize prev and node to be deleted 
        prev = head
        now = head.next
  
        while (prev != None and now != None): 
          
        # Change next link of previous node 
            prev.next = now.next
  
        # Free memory 
            now = None
  
        # Update prev and node 
            prev = prev.next
            if (prev != None): 
                now = prev.next
                
        return self.head
```

### Question 11 
Two Linked Lists are identical when they have the same data and the arrangement of data is <br>
also the same. Write a function to check if the given two linked lists are identical. <br>
Examples: <br>
Input: <br>
a = 1 → 2 → 3 → 4 → 5 → None <br>
b = 1 → 2 → 3 → 4 → 5 → None <br>
Output: Identical <br>
Input: <br>
a = 1 → 2 → 3 → 4 → None <br>
b = 1 → 7 → 4 → 5 → None <br>
Output: Not Identical <br>

```python
class Node:
    
    def __init__(self, element=None, next=None):
        self.element = element
        self.next = next


class SingleLinkedList:
    
    def __init__(self, a):
        if type(a) is list:
            self.head = Node(a[0], None)
            tail = self.head
            for i in range(1, len(a)):
                new_node = Node(a[i], None)
                tail.next = new_node
                tail = new_node
        else:
            self.head = a
    
    def printList(self):
        temp = self.head
        while temp.next != None:
            print(temp.element, end=" → ")
            temp = temp.next
        print(temp.element)
        return
    
    def question11(self, listb):
        a = self.head
        b = listb.head
        while (a != None and b != None):
            if (a.element != b.element):
                return False
 
            # If we reach here, then a and b
            # are not null and their data is
            # same, so move to next nodes
            # in both lists
            a = a.next
            b = b.next
 
        # If linked lists are identical,
        # then 'a' and 'b' must be null
        # at this point.
        return (a == None and b == None)
```

```python
class Node:
    
    def __init__(self, element=None, next=None):
        self.element = element
        self.next = next


class SingleLinkedList:
    
    def __init__(self, a):
        if type(a) is list:
            self.head = Node(a[0], None)
            tail = self.head
            for i in range(1, len(a)):
                new_node = Node(a[i], None)
                tail.next = new_node
                tail = new_node
        else:
            self.head = a
    
    def printList(self):
        temp = self.head
        while temp.next != None:
            print(temp.element, end=" → ")
            temp = temp.next
        print(temp.element)
        return
    
    def question12(self, k):

        first = last = self.head
        for i in range(1, k):
            first = first.next

        null_checker = first 
        while null_checker.next:
            last = last.next
            null_checker = null_checker.next
        first.element, last.element = last.element, first.element
        return self.head
```
